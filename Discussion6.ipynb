### The process of throwing dices    
<big>
On one hand, we know that when I throw one dice, there are $\frac{1}{6}$ that I would get one specific number, say, 4.  
If we throw $N$ times of dices, and we count how many times we got a 4, $N_{got 4}$. Then we can also say that the probability we got a 4 is:  
    $$P(number = 4) = \frac{N_{got 4}}{N}$$
So let's simulate the process of throwing dices and find the cumulative count of the presence of each number. The purpose of this short program is to show you how probability works (how we got a 1/6).

```python
probabilities = [1/6] * 6
np.random.multinomial(1, probabilities, size = None)
```

#### The above result shows the dice landed on the number 2.

> Indented block



```python
trials = 10000

rolls = np.random.multinomial(1, probabilities, size=trials)
counts = np.zeros((6,trials)) # 6 rows, 1000 trials
totals = np.zeros(6)
for i, roll in enumerate(rolls):
    totals[np.where(roll == 1)[0]] += 1
    counts[:, i] = totals
```

```python
totals
```

```python
rolls
```

```python
totals / trials
```

#### Normalizing by the number of tosses, we get the probabilities for each run.

```python
x = np.arange(trials).reshape((1,trials)) + 1
estimates = counts / x
print(estimates[:,0])
print(estimates[:,1])
print(estimates[:,100])
print(estimates[:,trials-1])
```

```python
plt.figure(figsize=(16, 12))
for i in range(6):
    plt.plot(estimates[i, :], label=("P(die=" + str(i+1) +")"))

plt.axhline(y=0.16666, color='black', linestyle='dashed')
plt.legend()   
plt.show()
```

# Example with 3 consequential regions: 
   
   
 height of the region = 1/length

 weight of the region = 1/the number regions


### Equation to find the contribution by one region: 
height of the region * weight of the region
    


### Question: Suppose I take 3 consequential regions of length 3, 5 and 7. Of these, the first 2 include 1 and third doesn’t. Lets calculate the contribution of each region for x = 1. 

```python
length_of_region_1 = 3 
length_of_region_2 = 5 
length_of_region_3 = 7 

number_of_regions = 3

weight_of_each_region = 1/number_of_regions

height_of_region_1 = 1/length_of_region_1
contribution_of_region_1 = height_of_region_1 * weight_of_each_region

height_of_region_2 = 1/length_of_region_2
contribution_of_region_2 = height_of_region_2 * weight_of_each_region


#region 3 does not contain x = 1: 
#therefore: 
contribution_of_region_3 = 0 


print("Contribution of region 1", contribution_of_region_1)
print("Contribution of region 2", contribution_of_region_2)
print("Contribution of region 3", contribution_of_region_3)

numerator = contribution_of_region_1 + contribution_of_region_2 + contribution_of_region_3
#denominator is height of the region * weight of the region regardless of whether x = 1 is in the range or not
denominator = 1/9 + 1/15 + 1/21
final_prob_x_equals_one = numerator/denominator

print("Final Probability for x =1", final_prob_x_equals_one)


```
